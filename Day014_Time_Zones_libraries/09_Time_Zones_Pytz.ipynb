```python
import pytz
from datetime import datetime
import pandas as pd
```

```python
my_datetime = datetime(2021, 1, 14, 15, 30)
print(my_datetime)
```

```python
my_datetime.tzinfo
```

```python
print(my_datetime.tzinfo)
```

```python
my_timezone_info = datetime(2021, 1, 14, 15, 30, tzinfo = pytz.UTC)
print(my_timezone_info)
```

```python
print(my_timezone_info.tzinfo)
```

```python
tz_US = my_timezone_info.astimezone(pytz.timezone('US/Central'))
print(tz_US)
```

```python
# show all timezones
pytz.all_timezones
```

# ------

```python
# Converting between Timeznes with Pandas
tz_US_pandas = pd.to_datetime(tz_US)
tz_US_pandas
```

```python
tz_UTC = tz_US_pandas.tz_convert('UTC')
print(tz_UTC)
```
